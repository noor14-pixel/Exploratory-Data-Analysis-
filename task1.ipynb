```python
import pandas as pd
import numpy as np 
import seaborn as sn
```

```python
df = pd.read_csv('Iris.csv') 
```

```python
df
```

```python
df = df.drop(columns=['Id'])
```

```python
print("Shape of the dataset:", df.shape)
print("\nColumn names:", df.columns.tolist())
print("\nFirst five rows:")
print(df.head())
```

```python
print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set a theme
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="PetalLengthCm", y="PetalWidthCm", hue="Species", palette="Set1")
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
df.hist(figsize=(12, 8), edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.drop(columns=['Species']), orient='h', palette='Set2')
plt.title("Box Plots for Feature Columns")
plt.show()

```
